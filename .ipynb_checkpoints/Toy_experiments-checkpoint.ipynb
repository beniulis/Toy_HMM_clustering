# Create HMM with different states and generate sequences:

## Create 2 HMMs

```python
from sklearn.cluster import DBSCAN
from __future__ import division
import numpy as np
from hmmlearn import hmm
import matplotlib.pyplot as plt
```

## 2 start probabilities for HMMs 

```python
start_probability_1 = np.array([0.8, 0.1, 0.1]) 
start_probability_2 = np.array([0.05, 0.9, 0.05])
```

## 2 Transition probabilities for HMMs 

```python
transition_probability_1 = np.array([
  [0.8, 0.1, 0.1], # add noise here delta (noise strength)
  [0.1, 0.8, 0.1],
  [0.5, 0.4, 0.1]
])

transition_probability_2 = np.array([
  [0.05, 0.9, 0.05],
  [0.05, 0.9, 0.05],
  [0.05, 0.9, 0.05]
])
```

## 2 Multinomial Emmision probabilities with 3 symbols to emmit:

```python
emission_probability_1 = np.array([
  [0.8, 0.1, 0.1],
  [0.1, 0.8, 0.1],
  [0.1, 0.1, 0.8]
])

emission_probability_2 = np.array([
  [0.1, 0.1, 0.8],
  [0.05, 0.05, 0.9],
  [0.8, 0.1, 0.1]
])
```

## Compile the HMM models:

```python
model_1 = hmm.MultinomialHMM(n_components=3)
model_1.startprob_=start_probability_1
model_1.transmat_=transition_probability_1
model_1.emissionprob_=emission_probability_1

model_2 = hmm.MultinomialHMM(n_components=3)
model_2.startprob_=start_probability_2
model_2.transmat_=transition_probability_2
model_2.emissionprob_=emission_probability_2
```

## Examples of sequences generated by each model

### HMM 1

```python
np.hstack(model_1.sample(100)[0])
```

### HMM 2

```python
np.hstack(model_2.sample(100)[0])
```

## Generate 100 sequences from each HMM model and combine all to have 200 sequences in total 

```python
hmm_seqs = [] # Save generated sequences  
hmm_seq = [] # Save generated sequences in different format 
hmm_labels = [] # Save the labels from which the sequences were generated 

for i in range(100):

    m1 = model_1.sample(100)[0] # 0th element for the emmisions of the generated sequences
    m2 = model_2.sample(100)[0]

    hmm_seqs.extend(m1) 
    hmm_seq.append(np.hstack(m1))
    hmm_labels.append(0)

    hmm_seqs.extend(m2)
    hmm_seq.append(np.hstack(m2))
    hmm_labels.append(1)
```

```python
hmm_seqs
```

## Lengths of the sequences

```python
lens = list(np.repeat(100, 200))
```

## Fit large HMM using the generated sequences:

### Total number of states: sum of states from the models that generated the sequences

```python
large_HMM = hmm.MultinomialHMM(n_components=6,n_iter=100,tol=0.01)
large_HMM.fit(hmm_seqs,lens)
```

```python
plt.plot(large_HMM.monitor_.history)
plt.show()
```

```python
large_HMM.startprob_
```

```python
large_HMM.emissionprob_
```

## Feed the sequences through the trained HMM to find the expected values of the transition matrices:

```python
from pomegranate import HiddenMarkovModel,DiscreteDistribution
```

### Build the pomegranate HMM from the parameters of the hmmlearn one

```python
from pomegranate import *
```

```python
from pomegranate import HiddenMarkovModel,DiscreteDistribution

elems = 3 # Number of unique elements from the emission matrix
hidden_states = 6

emiss = []
for emis_matrices in range(hidden_states):
    qw = large_HMM.emissionprob_[emis_matrices,:].tolist()
    s = DiscreteDistribution({ j : qw[j] for j in range(elems)})
    emiss.append(s)

model = HiddenMarkovModel.from_matrix(large_HMM.transmat_, emiss, large_HMM.startprob_)
model.bake()
```

```python
plt.imshow(large_HMM.transmat_)
plt.colorbar()
plt.show()
```

## Norm1 and Norm2 are the normalised induced transition matrices for the 2 sequences 

## Calculate the distances between the 2 sequences

```python
def bha(p,q):
    return -np.log(np.sum(np.sqrt(p*q)))
```

```python
bha(norm1[0],norm1[0])
```

### Affinity for all the rows

```python
np.sqrt(norm1*norm2).sum(1)
```

### Total affinity

```python
np.sqrt(norm1*norm2).sum(0).sum()
```

### Mean Affinity

```python
np.sqrt(norm1*norm2).sum()/6
```

### Final Distance

```python
-np.log(np.sqrt(norm1*norm2).sum()/6)
```

## Average affinity between the rows for the distance between the 2 transition matrices

```python
def bha_dist(p,q):
    return -np.log(np.sqrt(p*q).sum()/len(p))
```

```python
bha_dist(norm1,norm2)
```

## Calculate the distance between all the sequences

```python
def tot_bha_dist(seq1,seq2):
    
    tr1 = model.forward_backward(seq1)[0][0:6,0:6]
    tr2 = model.forward_backward(seq2)[0][0:6,0:6]
    
    norm1 = tr1/tr1.sum(axis=1,keepdims=1)
    norm2 = tr2/tr2.sum(axis=1,keepdims=1)

    return -np.log(np.sqrt(norm1*norm2).sum()/len(norm1))
```

## Same sequence distance should be 0

```python
tot_bha_dist(hmm_seq[1],hmm_seq[1])
```

### Calculate distance for all the sequences

```python
n = len(hmm_seq)
dists = np.zeros((n,n))
for i in range(n):
    for j in range(n):
        dists[i][j] = tot_bha_dist(hmm_seq[i],hmm_seq[j])
dists[dists < 0] = 0
```

```python
plt.hist(dists)
plt.show()
```

## The range of eps values in the DBSCAN algorithm should not be larger than the maximum distance between the sequences

```python
from sklearn import metrics
eps_range = np.arange(0.0001,np.amax(dists),0.0005) # Use the maximum values of the dists as upper limit of the eps values
min_samp_range = range(100) # Using half of the total number of sequences as an upper limit for the min samples range

sils = np.zeros((len(min_samp_range),len(eps_range)))
index = 0

for epes in eps_range:
    for min_samp in min_samp_range:
        clustering = DBSCAN(eps=epes, min_samples=min_samp,metric='precomputed').fit(dists)
        if len(np.unique(clustering.labels_)) < 2:
            sils[min_samp][index] = 0
        else:
            sils[min_samp][index] = metrics.silhouette_score(dists, clustering.labels_,metric='precomputed')
    index += 1
result = np.where(sils == np.amax(sils))
listOfCordinates = list(zip(result[0], result[1]))
```

```python
epeses, samples = np.meshgrid(eps_range, min_samp_range)
```

```python
from matplotlib import cm
from matplotlib.ticker import LinearLocator

fig, ax = plt.subplots(subplot_kw={"projection": "3d"},figsize=[15,15])
surf = ax.plot_surface(epeses, samples, sils,cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
samp_max = min_samp_range[listOfCordinates[-1][0]]
eps_max = eps_range[listOfCordinates[-1][1]]
```

```python
clustering = DBSCAN(eps=eps_max, min_samples=samp_max,metric='precomputed').fit(dists)
```

```python
def tot_bha_dist(seq1,seq2):
    
    tr1 = model.forward_backward(seq1)[0][0:6,0:6]
    tr2 = model.forward_backward(seq2)[0][0:6,0:6]
    
    norm1 = tr1/tr1.sum(axis=1,keepdims=1)
    norm2 = tr2/tr2.sum(axis=1,keepdims=1)

    return -np.log(np.sqrt(norm1*norm2).sum()/len(norm1))
```

```python
def calc_distances_hmms(inputs):
    
    # inputs: 0th - 2nd elements: start, trans, emmis probs of HMM-Model 1 
    # inputs: 3rd - 5th elements: start, trans, emmis probs of HMM-Model 2 
    
    model_1 = hmm.MultinomialHMM(n_components=len(inputs[0]))
    model_1.startprob_=inputs[0]; model_1.transmat_=inputs[1]; model_1.emissionprob_=inputs[2]

    model_2 = hmm.MultinomialHMM(n_components=len(inputs[3]))
    model_2.startprob_=inputs[3]; model_2.transmat_=inputs[4]; model_2.emissionprob_=inputs[5]
    
    # Generate sequences from the 2 models:
    
    num_of_models = 2 # Number of HMM models
    num_sequences = 100 # Number of sequences to be generated 
    hmm_seqs = [] # Save generated sequences  
    hmm_seq = [] # Save generated sequences in different format 
    hmm_labels = [] # Save the labels from which the sequences were generated 

    for i in range(num_sequences):
        
        m1 = model_1.sample(100)[0] # 0th element for the emmisions of the generated sequences
        m2 = model_2.sample(100)[0]
        
        hmm_seqs.extend(m1) 
        hmm_seq.append(np.hstack(m1))
        hmm_labels.append(0)
        
        hmm_seqs.extend(m2)
        hmm_seq.append(np.hstack(m2))
        hmm_labels.append(1)
    
    # Fit a large HMM with twice the number of states as the original models
    
    large_HMM = hmm.MultinomialHMM(n_components=num_of_models*len(inputs[0]),n_iter=100,tol=0.01,random_state=100)
    
    lens = list(np.repeat(num_sequences, num_sequences*num_of_models)) # Lens for the HMM fitting 
    large_HMM.fit(hmm_seqs,)
    
    from pomegranate import HiddenMarkovModel,DiscreteDistribution
    
    k_states = 3 # Number of unique emissions
    hidden_states = num_of_models*len(inputs[0]) # Total number of hidden states
    
    # Create an emmision probability matrix using the discrete distribution using the pomegranate package 
    
    emiss = []
    for emis_matrices in range(hidden_states):
        qw = large_HMM.emissionprob_[emis_matrices,:].tolist()
        s = DiscreteDistribution({ j : qw[j] for j in range(k_states)})
        emiss.append(s)

    # Define the large HMM model in pomegranate
    
    model = HiddenMarkovModel.from_matrix(large_HMM.transmat_, emiss, large_HMM.startprob_)
    model.bake()
    
    # Pass each sequence through the large HMM model and find the distances between all the sequences using the tot_bha_dist function
    
    
    def tot_bha_dist(seq1,seq2): 

        tr1 = model.forward_backward(seq1)[0][0:hidden_states,0:hidden_states] # Output the induced transition matrix
        tr2 = model.forward_backward(seq2)[0][0:hidden_states,0:hidden_states] # Take the elements based on total number of hidden states  

        norm1 = tr1/tr1.sum(axis=1,keepdims=1) # Normalise the induced transitions to obtain the 
        norm2 = tr2/tr2.sum(axis=1,keepdims=1) # transition probabilities for both sequences

         # Return the bhat distance between the sequences
         # The final distance is the average affinity between all the transition probabilities
         # for each hidden state.
        
        return -np.log(np.sqrt(norm1*norm2).sum()/len(norm1))
    
    
    # Find the distances between all the sequences using the tot_bha_dist function
    
    n = num_of_models*num_sequences

    dists = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            dists[i][j] = tot_bha_dist(hmm_seq[i],hmm_seq[j])
    dists[dists < 0] = 0
    
    # Use DBSCAN to cluster datapoints using the dists
    
    from sklearn import metrics
    
    # Define the ranges over which to find the DBSCAN parameters: 

    eps_range = np.arange(0.0001,np.amax(dists),0.0005) # Use the maximum values of the dists as upper limit of the eps values
    min_samp_range = range(100) # Using half of the total number of sequences as an upper limit for the min samples range

    sils = np.zeros((len(min_samp_range),len(eps_range)))
    
    # Index value to store which element of the eps value
    
    index = 0

    for epes in eps_range:
        for min_samp in min_samp_range:
            clustering = DBSCAN(eps=epes, min_samples=min_samp,metric='precomputed').fit(dists)
                
            # If there is only 1 cluster (Cluster label [-1] for outliers)
            # sil index is updated as 0
            
            if len(np.unique(clustering.labels_)) < 2:
                sils[min_samp][index] = 0
            else:
                sils[min_samp][index] = metrics.silhouette_score(dists, clustering.labels_,metric='precomputed')
        index += 1
    
    # Result and list of coordinates are the values of min_samples and eps that maximise the sillhoutte index 
    
    result = np.where(sils == np.amax(sils))
    listOfCordinates = list(zip(result[0], result[1]))
    
    return listOfCordinates
```

```python
inputs = [start_probability_1,transition_probability_1,emission_probability_1,start_probability_2,transition_probability_2,emission_probability_2]
```

```python
calc_distances_hmms(inputs)
```
